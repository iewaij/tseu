# Molotov
Molotov is a strategy that emphsize the potential non-linear relationship between momentum and volatility. It enriches dataframe with momentum, volatility and fundamental factors.

```python
def build_molotov(df):
    """
    Build fundamental, value, momentum, volatility, and size features.
    """
    df.fillna({
        "att": 0,
        "che": 0,
        "chee": 0,
        "ceq": 0,
        "dlc": 0,
        "dltt": 0,
        "mib": 0,
        "pstk": 0,
    }, inplace=True)
    df.eval("""
        fna = ivst + ivao
        fnl = dltt + dlc + pstk
        nca = att - act - ivaeq
        ncl = ltt - lct - dltt
        ncoa = nca - ncl
        nfna = fna - fnl
        oa = att - che
        ol = att - dlc - dltt - mib - pstk - ceq
        wc = act - che - lct + dlc
        liquid_cur = act / lct
        market_pe = mcap / (nicon * 10 ** 6)
        market_mb = mcap / (ceq * 10 ** 6)
        market_evs = (mcap + (dlc + dltt + pstk + mib - chee) * 10 ** 6) / (revt * 10 ** 6)
        market_eveda = (mcap + (dlc + dltt + pstk + mib - chee) * 10 ** 6) / (ebitda * 10 ** 6)        
        profitability_npm = nicon / revt
        profitability_ptm = pi / revt
        profitability_gpm=(revt - cogs) / revt
        profitability_roe = nicon / ceq
        profitability_roic = (ebit * (nicon / pi)) / (dlc + dltt + teq)
        profitability_roe = nicon / ceq
        """, inplace=True)
    firm_factor = lambda df: df.assign(
        accruals_chnfa=df.nfna - df.nfna.shift(1) / df.att.shift(1),
        profitability_at=df.revt / ((df.oa - df.ol) + (df.oa - df.ol).shift(1)) / 2,
        profitability_ct=df.revt / df.att.shift(1),
        profitability_gp=(df.revt - df.cogs) / df.att.shift(1),
        hign_52w=1 - df.close/df.close.rolling(12).max(),
        mom_12m=df.close.shift(1)/df.close.shift(12) - 1,
        mom_6m=df.close.shift(1)/df.close.shift(6) - 1,
        mom_1m=df.close / df.close.shift(1) - 1,
        vol_12m=df.close.rolling(12).std(),
        vol_6m=df.close.rolling(6).std(),
        vol_3m=df.close.rolling(3).std(),
        )
    industry_factor=lambda df: df.assign(
        ind_mom_12m=df.mom_12m.mean(),
        ind_mom_6m=df.mom_6m.mean(),
        ind_mom_1m=df.mom_1m.mean(),
    )
    df = df.groupby(["gvkey"]).apply(firm_factor)
    df = df.groupby(["date", "sic"]).apply(industry_factor)
    return df.loc[df.prccd>5, [
        "liquid_cur",
        "market_pe",
        "market_mb",
        "market_evs",
        "market_eveda",
        "accruals_chnfa",
        "profitability_ct",
        "profitability_gp",
        "profitability_npm",
        "profitability_ptm",
        "profitability_gpm",
        "profitability_roe",
        "profitability_roic",
        "hign_52w",
        "mom_12m",
        "mom_6m",
        "mom_1m",
        "vol_12m",
        "vol_6m",
        "vol_3m",
        "ind_mom_12m",
        "ind_mom_6m",
        "ind_mom_1m",
        "mcap",
        ]].dropna(subset=["mcap", "mom_12m"])
```

```python
def discipline_features(features):
    features = features.groupby("date").rank(pct=True)
    return features.fillna(0.5)
```

```python
def build_labels(universe, months=3, method="multi", thres=0.1, extreme=False):
    """
    extreme: Retain the 20% values that are the smallest and the 20% that are the largest.
    """
    log_ret = universe.close.groupby("gvkey").transform(lambda x: np.log(x.shift(-months)) - np.log(x)).dropna()
    y = log_ret.groupby("date").transform(lambda x: (x - x.median()).clip(-1, 1))
    
    if extreme:
        y = y.groupby("date", group_keys=False).apply(lambda x: x[(x < x.quantile(0.2)) | (x > x.quantile(0.8))])

    if method == "multi":
        thres = 0.1
        y_multi = pd.Series(0, index=y.index)
        y_multi.where(y > -thres, -1, inplace=True)
        y_multi.where(y < thres, 1, inplace=True)
        return y_multi
```

```python
def build_train_test(universe, train_start="2002-01-01", train_end="2012-01-01", test_start="2012-01-01", test_end="2016-01-01"):
    molotov_features = build_molotov(universe)
    features = discipline_features(molotov_features)
    labels = build_labels(universe)
    idx = labels.index.intersection(features.index)
    X = features.reindex(idx)
    y = labels.reindex(idx)
    X_train = X.xs(slice(train_start, train_end), level="date", drop_level=False)
    X_test = X.xs(slice(test_start, test_end), level="date", drop_level=False)
    y_train = y.xs(slice(train_start, train_end), level="date", drop_level=False)
    y_test = y.xs(slice(test_start, test_end), level="date", drop_level=False)
    return X_train, y_train, X_test, y_test
```

```python
def frame_to_signals(X, estimator):
    short_array = -1 * estimator.predict_proba(X)[:, 0]
    long_array = estimator.predict_proba(X)[:, -1]
    short_signal = pd.Series(short_array, X.index)
    long_signal = pd.Series(long_array, X.index)
    return short_signal, long_signal
```

```python
def noise_to_belief(signal, n):
    return signal.groupby("date", group_keys=False).apply(lambda x: x.sort_values(ascending=False).head(n))
```

```python
def monthly_to_daily(signal, months, weight):
    daily_signals = []
    for index, value in signal.iteritems():
        gvkey = index[0]
        predict_date = pd.to_datetime(index[1])
        start = predict_date + DateOffset(days=1)
        end = predict_date + DateOffset(months=months)
        date_range = pd.date_range(start, end, freq="D", name="date")
        daily_idx = pd.MultiIndex.from_tuples([(gvkey, date) for date in date_range], names=["gvkey", "date"])
        if weight == "equal":
            daily_signal = pd.Series(np.sign(value), daily_idx)
        elif weight == "value":
            daily_signal = pd.Series(value, daily_idx)
        daily_signals.append(daily_signal)
    return pd.concat(daily_signals)
```

```python
def signals_to_positions(short_signal, long_signal, short_n=50, long_n=50, months=3, weight="equal"):
    short_belief = noise_to_belief(short_signal, short_n)
    long_belief = noise_to_belief(long_signal, long_n)
    short_daily = monthly_to_daily(short_belief, months, weight)
    long_daily = monthly_to_daily(long_belief, months, weight)
    short_position = short_daily.groupby(["gvkey", "date"]).agg("sum")
    long_position = long_daily.groupby(["gvkey", "date"]).agg("sum")
    neutral_position = pd.concat([short_position, long_position]).groupby(["gvkey", "date"]).agg("sum")
    return short_position, long_position, neutral_position
```

```python
def position_to_margin(data, position, tx=-0.001, method="log"):
    merged = pd.merge(position.rename("position"), data.close, how="left", left_index=True, right_index=True)
    merged["close"] = merged.close.groupby("gvkey").fillna(method="ffill")
    merged["tx"] = np.where(position.shift(1) != position, tx, 0)
    if method == "log":
        merged["margin"] = merged.close.groupby("gvkey").transform(lambda x: np.log(x / x.shift(1))).fillna(0)
        margin = merged.position * merged.margin + merged.tx
        margin = margin.droplevel("gvkey").resample("D").mean().cumsum()
    elif method == "percent":
        merged["margin"] = merged.close.groupby("gvkey").transform(lambda x: x.pct_change(1)).fillna(0)
        margin = merged.position * merged.margin + merged.tx
        margin = 1 + margin.droplevel("gvkey").resample("D").mean().cumsum()
    return margin
```

```python
def sharpe_ratio(data, position):
    merged = pd.merge(position.rename("position"), data.close, how="left", left_index=True, right_index=True)
    merged["close"] = merged.close.groupby("gvkey").fillna(method="ffill")
    merged["tx"] = np.where(position.shift(1) != position, -0.001, 0)
    merged["margin"] = merged.close.groupby("gvkey").transform(lambda x: x.pct_change(1)).fillna(0)
    margin = merged.position * merged.margin + merged.tx
    margin = margin.droplevel("gvkey").resample("D").mean()
    return margin.mean() / margin.std() * np.sqrt(252)
```
