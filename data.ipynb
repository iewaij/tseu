```python
import pandas as pd
import pandas_datareader.data as web
import pandas_datareader as pdr
from datetime import timedelta, datetime, date
import requests_cache
import wrds
```

# Get Data

## `pandas_datareader`

```python
pdr.get_data_yahoo("ADS.DE")
```

```python
def get_data(symbols, start=datetime(2001, 1, 1), end=date.today()):
    expire_after = timedelta(days=1)
    session = requests_cache.CachedSession(
        cache_name="cache", backend="sqlite", expire_after=expire_after
    )
    data = web.DataReader(symbols, "yahoo", start, end, session=session)
    return data
```

```python
get_data("ADS.DE")
```

## `wrds`

```python
db = wrds.Connection(wrds_username='iewaij')
```

```python
def get_data_wrds(symbol):
    with wrds.Connection(wrds_username="iewaij") as db:
        data = db.raw_sql(
            f"""SELECT datadate, cshtrd, ajexdi, prcod, prccd, prchd, prcld 
                FROM comp.g_secd
                WHERE symbol= '{symbol}'
                AND datadate >= "2018-01-01"::date
            """,
            date_cols=["datadate"],
        )
    return data
```
